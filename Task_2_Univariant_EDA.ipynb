```python
# ===============================
# Step 1: Install & Import Libraries
# ===============================
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seaborn style
sns.set(style="whitegrid")
```

```python
# ===============================
# Step 2: Load Dataset
# ===============================
df = pd.read_csv("Mall_Customers.csv")
print("First 5 rows:\n", df.head())
```

```python
# ===============================
# Step 3: Data Inspection
# ===============================
print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nStatistical Summary:")
print(df.describe())
```

```python
# ===============================
# Step 4: Identify Variable Types
# ===============================
# Categorical: Gender
# Continuous: Age, Annual Income (k$), Spending Score (1-100)

# ===============================
# Step 5: Univariate Analysis â€“ Categorical Data
# ===============================

# Bar Chart - Gender Frequency
plt.figure(figsize=(5,4))
sns.countplot(data=df, x='Genre', palette='pastel')
plt.title("Bar Chart: Gender Distribution")
plt.show()
```

```python
# Pie Chart - Gender Proportion
gender_counts = df['Genre'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=['pink','lightblue'])
plt.title("Pie Chart: Gender Share")
plt.show()
```

```python
# Histogram - Age
plt.figure(figsize=(6,4))
sns.histplot(df['Age'], bins=15, kde=False, color='skyblue')
plt.title("Histogram: Age Distribution")
plt.show()
```

```python
# Density Plot - Annual Income
plt.figure(figsize=(6,4))
sns.kdeplot(df['Annual Income (k$)'], shade=True, color='orange')
plt.title("Density Plot: Annual Income")
plt.show()
```

```python
# Rug Plot - Spending Score
plt.figure(figsize=(6,4))
sns.rugplot(df['Spending Score (1-100)'], height=0.05, color="red")
plt.title("Rug Plot: Spending Score")
plt.show()
```

```python
# Strip Plot - Annual Income
plt.figure(figsize=(6,4))
sns.stripplot(x=df['Annual Income (k$)'], color="green")
plt.title("Strip Plot: Annual Income")
plt.show()
```

```python
# Swarm Plot - Age by Gender
plt.figure(figsize=(6,4))
sns.swarmplot(x="Genre", y="Age", data=df, palette="Set2")
plt.title("Swarm Plot: Age by Gender")
plt.show()
```

```python
# Scatter Plot - Age vs Spending Score
plt.figure(figsize=(6,4))
sns.scatterplot(x="Age", y="Spending Score (1-100)", data=df, color="purple")
plt.title("Scatter Plot: Age vs Spending Score")
plt.show()
```

```python
# Line Plot - Annual Income sorted
df_sorted = df.sort_values("Annual Income (k$)").reset_index()
plt.figure(figsize=(6,4))
sns.lineplot(x=df_sorted.index, y="Annual Income (k$)", data=df_sorted, marker="o", color="blue")
plt.title("Line Plot: Annual Income Trend (Sorted by Customers)")
plt.xlabel("Customer Index (sorted)")
plt.ylabel("Annual Income (k$)")
plt.show()
# ===============================
# Step 7 & 8: Interpret and Document Observations
# ===============================
# Example insights (to write in report):
# - Bar/Pie Chart â†’ Slightly more Female customers than Male.
# - Age Histogram â†’ Most customers are between 30â€“40 years.
# - Income KDE â†’ Annual income is uniformly distributed, peaks around 60k.
# - Spending Score Rug/Strip â†’ Customers spread across the full 1â€“100 range.
# - Swarm Plot â†’ Younger age group is balanced across genders.
# - Scatter Plot â†’ Some younger customers have very high spending scores.
# - Line Plot â†’ No smooth trend, but shows variance in incomes across customers.
```
